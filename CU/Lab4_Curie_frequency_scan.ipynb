```python

import numpy as np
import matplotlib.pyplot as plt
```

```python
url = 'https://raw.githubusercontent.com/evgeny-kolonsky/Lab4/refs/heads/main/CU/frequency_scan_13.txt'
f, Vp, Vs_air, Vs1, Vs2 = np.loadtxt(url).T
f *= 1e3 # -> Hz
```

```python
plt.plot(f, Vs_air / Vp, '-o', label='air')
plt.plot(f, Vs1 / Vp, '-o', label='sample1')
plt.plot(f, Vs2 / Vp, '-o', label='sample2')
plt.xscale('log')
plt.grid()
plt.xlabel('f (kHz)')
plt.ylabel('Vs/Vp')
#plt.xlim(0, 1e4)
plt.legend()
```

```python
plt.plot(f, Vs1/Vs_air, '-o', label='sample1/air')
plt.plot(f, Vs2/Vs_air, '-o', label='sample2/air')
plt.xscale('log')
plt.grid()
plt.xlabel('f (kHz)')
plt.ylabel('Vs/Vair')
#plt.xlim(0, 1e4)
plt.legend()
```
