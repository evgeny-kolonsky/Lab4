```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
```

```python
filename = 'https://raw.githubusercontent.com/tphlabs/data/refs/heads/main/Lab4_Xray/2025.01_Bragg/Mo_LiF.txt'
theta, count = np.genfromtxt(filename, skip_header=3, unpack=True)
theta_rad = theta * np.pi / 180

peaks, _ = find_peaks(count, prominence=50)
```

```python
plt.plot(theta[peaks], count[peaks], 'x')
plt.plot(theta, count)
plt.yscale('log')
plt.xlabel('Theta crystal, grad')
plt.ylabel('Count rate, #/s')
plt.grid()
theta[peaks]
```
