$$
E = \acute E + \hbar \omega, \quad E = e U_A;
$$
$$
\acute E \ge 0; \quad \omega_{max} = \frac{E}{\hbar}
$$
$$
\lambda_{min} = \frac{2 \pi c}{\omega_{max} }
$$
therefore
$$
\hbar = \frac{e U_A \lambda_{min}}{2 \pi c}, \quad h = 2 \pi \hbar  
$$

Bragg's diffraction

$$
n \lambda = 2 d \sin \theta
$$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.constants import speed_of_light as c, elementary_charge as e, Planck as h
from scipy.stats import linregress

data = '''
# Cuprum radiation
# Ua [kV], theta_min [deg]
35 5.5
27 7
20 9
'''

Ua, theta_min = np.genfromtxt(data.splitlines(), unpack=True)
Ua *= 1e3
theta_min *= np.pi / 180
d = 2e-10 # atomic planes distance, m
lambda_min = 2 * d * np.sin(theta_min)

h_exp = e * Ua * lambda_min / c

np.set_printoptions(precision=2)
print('Observations    [j/s] ^e34')
print(h_exp *1e34)
print('Reference value [j/s] ^e34:')
print(f'{h *1e34:.2f}')
```
