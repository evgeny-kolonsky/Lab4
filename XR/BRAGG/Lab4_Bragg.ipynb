```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
from scipy.optimize import curve_fit
```

```python
URL = 'https://raw.githubusercontent.com/evgeny-kolonsky/Lab4/refs/heads/main/XR/BRAGG/'
crystal = 'LF1'
for lamp in ['Mo1', 'Mo2', 'Mo3', 'Mo4']:
  filename = URL + f'{lamp}_{crystal}_d2.txt'
  theta, count = np.genfromtxt(filename, skip_header=3, unpack=True)
  theta_rad = theta * np.pi / 180
  peaks, _ = find_peaks(count, prominence=100)
  plt.plot(theta, count,  label=lamp)
  #plt.plot(theta[peaks], count[peaks], 'x')

plt.yscale('log')
plt.xlabel('Theta crystal, grad')
plt.ylabel('Count rate, #/s')
plt.grid()
plt.legend()
plt.xlim(5, 45)
```

## Crystal `LiF1` is performing better than `LiF2` or `LiF3`.


```python
lamp = 'Mo3'
for crystal in ['KB1', 'KB2']:
  filename = URL + f'{lamp}_{crystal}_d2.txt'
  theta, count = np.genfromtxt(filename, skip_header=3, usecols=[0, 1], unpack=True, encoding='latin1')
  theta_rad = theta * np.pi / 180
  peaks, _ = find_peaks(count, prominence=100)
  plt.plot(theta, count,  label=crystal)
  plt.plot(theta[peaks], count[peaks], 'x')

plt.yscale('log')
plt.xlabel('Theta crystal, grad')
plt.ylabel('Count rate, #/s')
plt.grid()
plt.legend()
plt.xlim(5, 45)
```
