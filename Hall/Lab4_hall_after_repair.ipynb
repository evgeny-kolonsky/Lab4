Repaired cart #2896 measurements

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
# Constants
d = 1e-3 # thikness, m
L = 16e-3 # length between ohmic contacts, m
W = 10e-3 # width, m
```

```python
# Sample 13 repaired cart #2896 measurements
```

```python
Uh_min, Uh_max = -144, -125 # mV
Ip = 29e-3 # A
data = '''
-271 -146
-243 -143
-210 -140
-182 -138
-154 -136
-120 -133
-94 -131
-72 -129
-52 -127
-19 -124
-6 -123
# polarity
273 -101
242 -103
207 -106
171 -109
145 -111
120 -113
82 -116
53 -118
22 -121
6 -122
'''

B, Uh = np.genfromtxt(data.splitlines()).T
B  *=-1e-3 # mT -> T
Uh *=1e-3 # mV -> V
plt.plot(B, Uh, '.')
plt.xlabel('B, T')
plt.ylabel('Uh, V')
plt.grid()

fit = linregress(B, Uh)
RH = fit.slope * d / Ip
plt.plot(B, fit.intercept + fit.slope *B, '--', label=rf'Fit $R_H ={RH*1e2:.4f}$ Ohm')
plt.legend()
```
